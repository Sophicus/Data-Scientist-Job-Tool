```python
def word_cloud_job_title(data, font_size = 40, title = '') :
    
    stopwords = ['etc','years', 'Etc','degree','skill','using','preferred','field',
                'based','related','including','Ability','experience','tools','will','working','process','Ecosystem','Work'
                'new','support','team','opportunities','data','new','make','closely','Data','Develop','It','good'
                'language','help','analysis','sourced','long','understand','Brush','Billed','language']
    for word in stopwords:
        if word in data:
            data=data.replace(word,"")
        
    # Generate a word cloud image
    wordcloud = WordCloud().generate(data)
    
    # Display the generated image:
    # the matplotlib way:
    #lt.imshow(wordcloud, interpolation='bilinear')
    #lt.axis("off")
    #fig = plt.gcf()
    #fig.set_size_inches(15,10)
    #plt.title(title, fontsize = 24)
    
    # lower max_font_size
    wordcloud = WordCloud(max_font_size=font_size).generate(data)
    plt.figure()
    plt.imshow(wordcloud, interpolation="bilinear")
    plt.axis("off")
    fig = plt.gcf()
    fig.set_size_inches(15,10)
    plt.title(title, fontsize = 24)
    plt.show()
```

```python
### Data_analyst responsibilites
word_cloud_job_title(data_analyst_resp, title = 'data_analyst_responsibility')
```

```python
### Data_analyst skill
word_cloud_job_title(data_analyst_skill, title = 'data_analyst_skill')
```
