```python
# Importing the Libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.cross_validation import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
# Importing the dataset
dataset = pd.read_csv('Resources/Salary_Data.csv')
X = dataset.iloc[:, :-1].values
Y = dataset.iloc[:, 1].values # Dependent Variable/Target Values
```

```python
# Splitting the Dataset into Training Set and Test Set
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 1/2, random_state = 0)
print(Y_test)
```

```python
regressor = LinearRegression()
regressor.fit(X_train, Y_train)
```

```python
# Predicting the Test Set Results
y_pred = regressor.predict(X_test)
y_pred
```

```python
# Visualizing the Training Set Results
plt.scatter(X_train, Y_train, color = 'red')
#plt.plot(X_train, regressor.predict(X_train), color = 'blue')
plt.scatter(X_train,regressor.predict(X_train),color="blue")
plt.title('Salary Prediction on a Training Set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.savefig("Output\Training_Set_ML.png")
plt.show()
```

```python
plt.scatter(X_test,Y_test,color="red")
plt.scatter(X_test,y_pred,color="blue")
plt.title("Salary Prediction on a Test DataFrame")
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.savefig("Output\Test_Set_ML.png")
plt.show()
```
