```python
from pytrends.request import TrendReq
import pandas as pd
import matplotlib.pyplot as plt

pytrends = TrendReq(hl='en-US', tz=360)
```

```python
kw_list = ["big data", 'data science', 'machine learning']
pytrends.build_payload(kw_list, cat=0, timeframe='today 5-y', geo='US', gprop='images')
```

```python
df = pytrends.interest_over_time()
```

```python
df.head()
```

```python
df.plot()
_ = plt.plot
```

```python
df['big data'].plot()
plt.title('Big Data Google Searches')
_ = plt.plot
```

```python
df['machine learning'].plot()
plt.title('Machine Learning Google Searches')
_ = plt.plot
```

```python
df['data science'].plot()
plt.title('Data Science Google Searches')
_ = plt.plot
```
